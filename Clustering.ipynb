##### Clustering ######

Q1.What is unsupervised learning in the context of machine learning?
==>In the context of machine learning, unsupervised learning is a type of learning where algorithms identify patterns and structures in unlabeled data without any explicit guidance or supervision. Unlike supervised learning, which relies on labeled data to train models, unsupervised learning aims to discover hidden relationships and groupings within data.

Q2.How does K-Means clustering algorithm work?
==>K-means clustering is a machine learning algorithm that groups similar data points into clusters. It works by iteratively assigning each data point to the nearest centroid, which is the average of all data points in a cluster, and then recalculating the centroid's position based on the new assignments. This process continues until the cluster assignments stabilize, meaning data points are no longer moving between clusters.

Q3.Explain the concept of a dendrogram in hierarchical clustering?
==>In hierarchical clustering, a dendrogram is a tree-like diagram that visually represents the hierarchy of clusters formed during the clustering process. It shows how data points or clusters are merged together based on their similarity or distance. The height of the dendrogram indicates the level at which clusters are combined, with shorter lines signifying higher similarity.

Q4.What is the main difference between K-Means and Hierarchical Clustering?
==>Hierarchical Clustering vs K-Means Clustering: How do the ...The main difference between K-means and hierarchical clustering lies in how they handle the number of clusters and the resulting structure of the clusters. K-means requires a pre-specified number of clusters (K), while hierarchical clustering builds a hierarchy of clusters without a predetermined number. K-means is computationally faster and simpler, but may not capture complex cluster structures, whereas hierarchical clustering is more flexible and intuitive but can be computationally expensive.

Q5.What are the advantages of DBSCAN over K-Means?
==>DBSCAN offers several advantages over K-Means, including its ability to handle clusters of any shape, its robustness to noise and outliers, and its capacity to identify clusters of varying densities. Unlike K-Means, which requires pre-defining the number of clusters and assumes spherical shapes, DBSCAN automatically determines the number of clusters and can identify clusters of irregular forms.

Q6.When would you use Silhouette Score in clustering?
==>The Silhouette Score is primarily used in clustering analysis to evaluate the quality of clusters and help determine the optimal number of clusters for a given dataset. It measures how similar a data point is to its own cluster compared to other clusters. A higher Silhouette Score generally indicates better clustering, with scores closer to 1 suggesting strong clustering and scores close to -1 suggesting overlapping clusters.

Q7.What are the limitations of Hierarchical Clustering?
==>Hierarchical Clustering, while powerful, has limitations. Primarily, it suffers from high computational complexity and memory requirements, especially with large datasets. It's also sensitive to outliers and noise, which can distort the resulting clusters. Additionally, the merging and splitting decisions are irreversible, potentially leading to suboptimal cluster structures.

Q8.Why is feature scaling important in clustering algorithms like K-Means?
==>Feature scaling is crucial in clustering algorithms like K-Means because it prevents features with larger magnitudes from dominating the clustering process, ensuring that all features are considered equally important. Without scaling, a large-magnitude feature could disproportionately influence distance calculations, leading to skewed cluster formations.
* Distance-Based Algorithms:
K-Means, like many clustering algorithms, relies on distance metrics (like Euclidean distance) to determine the similarity between data points.
* Dominant Features:
If features have significantly different scales (e.g., income in thousands vs. age in years), the larger-magnitude features will have a greater impact on the calculated distances.
* Biased Clustering:
This can lead to a scenario where clusters are formed based on the dominant feature, even if other features are equally important in defining the underlying groupings.

Q9.How does DBSCAN identify noise points?
==>DBSCAN handles noisy data points by classifying them as "noise" rather than assigning them to a cluster, effectively ignoring them for the purpose of clustering. It identifies clusters based on the density of data points, so points that are isolated or sparsely distributed are not part of any cluster and are labeled as noise. This makes DBSCAN robust to outliers and irregularly shaped clusters.

Q10.Define inertia in the context of K-Means.
==>K-Means: Inertia
It is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster. A good model is one with low inertia AND a low number of clusters ( K ). However, this is a tradeoff because as K increases, inertia decreases.

Q11.What is the elbow method in K-Means clustering?
==>The elbow method is a heuristic used in k-means clustering to determine the optimal number of clusters (k). It involves plotting the within-cluster sum of squares (WCSS) for a range of k values and visually identifying the "elbow" point on the graph. The elbow point represents the point where the rate of decrease in WCSS sharply changes, suggesting that adding more clusters beyond this point yields diminishing returns.

Q12.Describe the concept of "density" in DBSCAN.
==>In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), "density" refers to the concentration of data points within a certain radius around a given point. DBSCAN uses this concept to identify clusters as regions of high density separated by areas of lower density. Essentially, clusters are formed by grouping points that are closely packed together and are considered to be part of the same cluster if they are connected through a chain of dense regions.

Q13.Can hierarchical clustering be used on categorical data?
==>Yes, hierarchical clustering can be used on categorical data. However, it requires using appropriate distance metrics that are suited for categorical data, such as Gower's distance or the Hamming distance. Traditional hierarchical clustering algorithms like agglomerative clustering can be adapted for categorical data by employing these specialized distance functions.

Q14.What does a negative Silhouette Score indicate?
==>Silhouette coefficients close to + 1 imply that the sample is far from neighboring clusters. A value of 0 means that the sample is on or very close to the decision boundary between two neighboring clusters, and negative values suggest that the samples may have been allocated to the wrong cluster.

Q15.Explain the term "linkage criteria" in hierarchical clustering?
==>In hierarchical clustering, "linkage criteria" defines how the distance between two clusters is calculated when merging them into a larger cluster. It essentially dictates the "glue" that holds clusters together during the hierarchical merging process. Different linkage criteria can lead to different cluster structures and shapes.
1. What it is:
Linkage criteria specify how to measure the distance between two clusters.
This distance is used to determine which clusters to merge at each step of the hierarchical clustering algorithm.
The choice of linkage criterion significantly impacts the resulting cluster structure.
2. Common Linkage Criteria:
Single Linkage (Minimum):
Measures the distance between the closest pair of points, one from each cluster.
Good for identifying elongated or "chain-like" clusters.
Sensitive to outliers, as a single outlier can bridge distant clusters.
3. Why it matters:
Different linkage criteria lead to different cluster shapes and structures.
The choice of linkage criterion depends on the characteristics of the data and the desired clustering outcome.
Understanding linkage criteria is crucial for interpreting the results of hierarchical clustering.

Q16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
==>K-means clustering, a popular clustering algorithm, struggles with data containing varying cluster sizes and densities because it's based on the assumption that all clusters are roughly spherical and of similar size. When clusters differ significantly in size or density, the algorithm can misclassify data points and produce poor results.
*K-means assumes spherical clusters:
K-means aims to find the centroids of spherical clusters by minimizing the distance of data points to their closest centroid. This assumption works well when clusters are roughly spherical and have similar densities.
*Uneven cluster sizes:
If some clusters are significantly larger than others, the algorithm might be pulled towards the larger cluster, leading to the misclassification of data points from smaller clusters.
*Density variations:
When there are areas of high and low density within the data, K-means can struggle to identify distinct clusters. It might mistakenly group points that are close together in dense regions, while failing to separate points from sparse regions.
*Outliers:
K-means is also highly sensitive to outliers. Outliers can significantly influence the position of cluster centroids, leading to distorted cluster assignments.

Q17.What are the core parameters in DBSCAN, and how do they influence clustering?
==>DBSCAN, a density-based clustering algorithm, relies on two core parameters: Epsilon (ε) and Minimum Points (MinPts). These parameters directly influence how DBSCAN identifies clusters and noise. Epsilon defines the radius around a point, and MinPts determines the minimum number of points within that radius required for a point to be considered a core point, indicating a dense area.
Smaller ε:
Leads to more, smaller clusters. Points within a shorter distance are grouped together, potentially forming more isolated clusters.
Larger ε:
Results in fewer, larger clusters. Points are grouped together over a greater distance, potentially merging clusters that would be separate with a smaller ε.
Balancing Granularity and Performance:
Choosing the right ε is crucial for balancing the granularity of the clusters (how many distinct groups) and the algorithm's performance.

Q18.How does K-Means++ improve upon standard K-Means initialization?
==>Although the initialization in K-means++ is computationally more expensive than the standard K-means algorithm, the run-time for convergence to optimum is drastically reduced for K-means++. This is because the centroids that are initially chosen are likely to lie in different clusters already.

Q19.What is agglomerative clustering?
==>Agglomerative clustering is a hierarchical clustering technique that uses a "bottom-up" approach, starting with each data point as a separate cluster and then iteratively merging the most similar clusters until a single cluster remains. This process creates a hierarchy of clusters, often visualized as a dendrogram.

Q20.What makes Silhouette Score a better metric than just inertia for model evaluation?
==>Silhouette Score is a superior metric for clustering model evaluation compared to inertia because it considers both intra-cluster cohesion (how tightly packed points are within a cluster) and inter-cluster separation (how well clusters are separated from each other). In contrast, inertia primarily focuses on within-cluster compactness, potentially overlooking the overall clustering structure and separation.

```python
#Q8.Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means.
from time import time

from sklearn import metrics
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler


def bench_k_means(kmeans, name, data, labels):
    """Benchmark to evaluate the KMeans initialization methods.

    Parameters
    ----------
    kmeans : KMeans instance
        A :class:`~sklearn.cluster.KMeans` instance with the initialization
        already set.
    name : str
        Name given to the strategy. It will be used to show the results in a
        table.
    data : ndarray of shape (n_samples, n_features)
        The data to cluster.
    labels : ndarray of shape (n_samples,)
        The labels used to compute the clustering metrics which requires some
        supervision.
    """
    t0 = time()
    estimator = make_pipeline(StandardScaler(), kmeans).fit(data)
    fit_time = time() - t0
    results = [name, fit_time, estimator[-1].inertia_]

    # Define the metrics which require only the true labels and estimator
    # labels
    clustering_metrics = [
        metrics.homogeneity_score,
        metrics.completeness_score,
        metrics.v_measure_score,
        metrics.adjusted_rand_score,
        metrics.adjusted_mutual_info_score,
    ]
    results += [m(labels, estimator[-1].labels_) for m in clustering_metrics]

    # The silhouette score requires the full dataset
    results += [
        metrics.silhouette_score(
            data,
            estimator[-1].labels_,
            metric="euclidean",
            sample_size=300,
        )
    ]

    # Show the results
    formatter_result = (
        "{:9s}\t{:.3f}s\t{:.0f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}\t{:.3f}"
    )
    print(formatter_result.format(*results))
```

```python
#Q9.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Calculate silhouette scores for k=2 to 5
silhouette_scores = []
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)  # added n_init for warning
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(range(2, 6), silhouette_scores, color='skyblue')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Number of Clusters')
plt.xticks(range(2, 6))
plt.grid(axis='y', linestyle='--')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#Q11.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from matplotlib.colors import ListedColormap
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
# Create a meshgrid for plotting decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
Z_kmeans = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z_kmeans = Z_kmeans.reshape(xx.shape)
# Define color map for cluster zones
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])

# Plot decision boundaries for KMeans
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z_kmeans, cmap=cmap_light, alpha=0.6)

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=50, edgecolor='k', cmap='viridis')

# Plot the cluster centers
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='x')

plt.title('KMeans Clustering - Cluster Boundaries')
plt.show()
```

```python
#Q14.Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
X = breast_cancer.data

# Determine the range of K values to test
range_k = range(2, 7)

# Calculate inertia (within-cluster sum of squares) for each K
inertia_values = []
for k in range_k:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

# Create a line plot to visualize inertia values
plt.plot(range_k, inertia_values, marker='o')
plt.title('Inertia for Different K Values')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

```python
#Q16.Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding noise).
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler

# 1. Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target

# 2. Data Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=10)
dbscan.fit(X_scaled)

labels = dbscan.labels_

# 4. Count Clusters
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters:", n_clusters_)
print("Estimated number of noise points:", n_noise_)
```

```python
#Q17.Generate synthetic data with make_blobs and apply KMeans. Then plot the cluster centers on top of the data points.
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np

# 1. Generate synthetic data using make_blobs
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=42)

# 2. Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)
kmeans_labels = kmeans.labels_

# 3. Extract cluster centers
cluster_centers = kmeans.cluster_centers_

# 4. Plot the data points and cluster centers
plt.figure(figsize=(8, 6))

# Scatter plot of data points
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, s=50, cmap='viridis')

# Scatter plot of cluster centers
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=200, c='red', marker='X', label='Cluster Centers')

plt.title('K-Means Clustering on Synthetic Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)

plt.show()
```

```python
#Q20.Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D scatter plot.
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
print(iris.keys())
import seaborn as sns

# Rename classes using the iris target names
iris.frame["target"] = iris.target_names[iris.target]
_ = sns.pairplot(iris.frame, hue="target")
import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn.decomposition import PCA

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(iris.data)
scatter = ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=iris.target,
    s=40,
)

ax.set(
    title="First three PCA dimensions",
    xlabel="1st Eigenvector",
    ylabel="2nd Eigenvector",
    zlabel="3rd Eigenvector",
)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

# Add a legend
legend1 = ax.legend(
    scatter.legend_elements()[0],
    iris.target_names.tolist(),
    loc="upper right",
    title="Classes",
)
ax.add_artist(legend1)

plt.show()
```
